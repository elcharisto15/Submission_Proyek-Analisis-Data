## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana kondisi cuaca (suhu, kelembapan dan kecepatan angin) memengaruhi penyewaan sepeda?

```python
# rata rata jumlah sewa berdasarkan kondisi cuaca menggunakan plot
plt.figure(figsize=(10, 6))
sns.barplot(x='weathersit', y='cnt', data=hour_df, color='yellow')
plt.title('Rata-rata Jumlah Sewa Sepeda Berdasarkan Kondisi Cuaca')
plt.xlabel('Kondisi Cuaca (1: Cerah, 2: Berkabut, 3: Hujan Ringan, 4: Hujan Lebat)')
plt.ylabel('Jumlah Sewa')
plt.show()

# Menghitung korelasi
korelasi = hour_df['cnt'].corr(hour_df['weathersit'])
print("Korelasi antara cnt dan weathersit:", korelasi)
```

**Insight:**
- Dilihat dari grafik jika jumlah sewa sepeda paling tinggi yaitu ketika kondisi cuaca cerah
- Dari grafik menurun berdasarkan kondisi cuaca dari cerah sampai ke hujan lebat

### Pertanyaan 2: Bagaimana pola rata rata sewa sepeda berdasarkan tiap bulan?

```python
perbulan = day_df.groupby('mnth')['cnt'].mean().reset_index()
bulan = ['Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni', 'Juli', 'Agustus', 'September', 'Oktober', 'November', 'Desember']
perbulan['mnth'] = perbulan['mnth'].apply(lambda x: bulan[x-1])
plt.figure(figsize=(10, 5))
sns.barplot(data=perbulan, x='mnth', y='cnt')
plt.title('Rata-rata Sewa tiap bulan', fontsize=16)
plt.xlabel('Bulan')
plt.ylabel('Jumlah Rata-rata Sewa Sepeda')
plt.show()
```

**Insight:**
- Dilihat dari grafik jika dibulan juni dan september memiliki rata rata sewa sepeda yang sama sehingga dapat dilihat bahwa rata rata sewa tertinggi terdapat pada bulan tersebut.
